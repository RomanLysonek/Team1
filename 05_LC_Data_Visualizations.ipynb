```python
x_vals = range(1, 101)
y_vals = [2*x for x in x_vals]
y_vals2 = [4*x for x in x_vals]

fig, ax = plt.subplots() #Creates an empty figure (i.e. "canvas") and axes (i.e. the actual plot)

## The following lines are now direct methods of the instantiated "ax" object ##
ax.plot(x_vals, y_vals)
ax.plot(x_vals, y_vals2)
ax.set_title('My Title') #notice that the method is called "set_title()"
ax.set_ylabel('y values')
ax.set_xlabel('x values')
ax.legend(['y = 2x', 'y = 4x'])
ax.grid()

plt.show()
```

```python
x_vals = range(1, 101)
y_vals = [2*x for x in x_vals]
y_vals2 = [np.sin(x) for x in x_vals]

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharex=True)
#Creates an empty figure (i.e. "canvas") with a 2x1 grid of sub-plots.

#Each subplot / axes object can be treated differently
ax1.plot(x_vals, y_vals)
ax1.set_title('y = 2x')
ax1.set_ylabel('y values')

ax2.plot(x_vals, y_vals2)
ax2.set_title('y = sin(x)')
ax2.set_ylabel('y values')
ax2.set_xlabel('x values')

plt.tight_layout() #Automatically adjust subplot parameters to give specified padding.
plt.show()
```

```python
import math
import random

# prepare random data
n_balls = 50
x = [random.triangular() for i in range(n_balls)]
y = [random.gauss(0.5, 0.15) for i in range(n_balls)]
colors = [random.randint(1, 4) for i in range(n_balls)]
areas = [math.pi * random.randint(5, 15)**2 for i in range(n_balls)]

# draw the figure
plt.figure()
plt.scatter(x, y, s=areas, c=colors, alpha=0.75)
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.savefig('balls.png')
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y) #further details in Numpy lectures
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=5, cmap=cm.coolwarm)
plt.show()
```
